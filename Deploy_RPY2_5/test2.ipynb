```python
import json
import pandas as pd
def anova_df_to_json(df):
    # NA 값을 0으로 대체
    df = df.fillna(0)

    # 데이터프레임을 JSON 형식으로 변환
    anova_list = []
    for response_type, group in df.groupby("response_type"):
        sources = {}
        for _, row in group.iterrows():
            sources[row["result_type"]] = {
                "df": row["DF"],
                "sum_sq": row["sumsq"],
                "mean_sq": row["meansq"],
                "F value": row["F_VALUE"],
                "Pr(>F)": row["PR(>F)"]
            }
        anova_list.append({
            "response_type": response_type,
            "source": sources
        })

    anova_json = json.dumps({"anova": anova_list}, indent=4)
    return anova_json
```

```python
def anova_df_to_list(df):
    # NA 값을 0으로 대체
    df = df.fillna(0)

    # 데이터프레임을 리스트 형식으로 변환
    anova_list = []
    
    for response_type, group in df.groupby("response_type"):
        sources = {}
        for _, row in group.iterrows():
            sources[row["source"]] = {
                "df": row["DF"],
                "sum_sq": row["sumsq"],
                "mean_sq": row["meansq"],
                "F value": row["F_VALUE"],
                "Pr(>F)": row["PR(>F)"]
            }
        anova_list.append({
            "response_type": response_type,
            "source": sources
        })

    return anova_list
```
